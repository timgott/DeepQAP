```python
from drlqap.taskgenerators import small_random_graphs
from drlqap.simplesolver import solve_qap_backtracking
```

```python
generator = small_random_graphs()
```

```python
qaps = generator.test_set()
```

```python
optimal_values = []
optimal_assignments = []

for i, qap in enumerate(qaps):
    value, assignment = solve_qap_backtracking(qap)
    print(i, value, assignment)
    optimal_values.append(value)
    optimal_assignments.append(assignment)
```

```python
with open("small_random_graphs_optimums.txt", "w") as f:
    f.writelines([str(v.item())+ '\n' for v in optimal_values])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(x=range(len(optimal_values)), y=optimal_values)
```

```python
mean = np.mean(optimal_values)
minimum = np.min(optimal_values)
maximum = np.max(optimal_values)
```

```python
print("Mean:", mean)
print("Min:", minimum)
print("Max:", maximum)
```
